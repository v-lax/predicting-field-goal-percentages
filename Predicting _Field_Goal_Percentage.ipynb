```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
season_stats=pd.read_csv('Seasons_Stats.csv')
season_stats.info()
```

```python
season_stats1=season_stats.drop(['Unnamed: 0','Year','blank2','blanl','Tm','Age',
                                 'Pos','OWS','DWS',
                                 'WS','WS/48','OBPM','DBPM','BPM'],
                                axis=1)
```

```python
player_sea_stats=season_stats1.groupby('Player')
player_sea_stats
        
    
```

```python
average_pts_per_season=player_sea_stats['FG%'].mean()
average_pts_per_season
```

```python
players=player_sea_stats.mean()
players
```

```python
legit_players=players.dropna()
legit_players
```

```python
legit_players.describe()
```

```python
legit_players.hist(figsize=(18,18));
```

```python
# it looks like the variables that have a normal distribution are 
# 2P%,AST%, DRB%, FG%, FT%, G,PER, TOV%,TRB%, TS%, USG%, VORP, eFG%
normal_legit_players=legit_players[['2P%','AST%','DRB%','FG%','FT%','G',"PER",'TOV%','TRB%','TS%','USG%','VORP','eFG%']]
data=normal_legit_players.rename(columns={'2P%':'TwoPPer','AST%':'ASTPer','DRB%':'DRBPer','FG%':'FGPer','FT%':'FTPer','TOV%':'TOVPer','TRB%':'TRBPer','TS%':'TSPer','USG%':'USGPer','eFG%':'eFGPer'})
data
```

```python
for column in ['TwoPPer','ASTPer','DRBPer','FTPer','G',"PER",'TOVPer','TRBPer','TSPer','USGPer','VORP','eFGPer']:
    plt.scatter(data[column],data.FGPer,label=column)
    plt.legend()
    plt.show()
```

```python
# so it looks like the only features that have somewhat of a linear
# relationship are 'VORP', 'PER','FT%'

# import libraries
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import statsmodels.stats.api as sms
```

```python
results=[['ind_var','r_squared','intercept','slope','p-value','normality']]
for index,val in enumerate(['TwoPPer','ASTPer','DRBPer','FTPer','G',"PER",'TOVPer','TRBPer','TSPer','USGPer','VORP','eFGPer']):
    print('NBA Player Season Data - Regression Analysis for formula: FGPer~'+val)
    print('-------------------------------------------------------------------')
    print()
    
    f='FGPer~'+val
    
    model= smf.ols(formula=f,data=data).fit()
    
    new_x= pd.DataFrame({val:[data[val].min(),data[val].max()]})
    predictions=model.predict(new_x)
    data.plot(kind='scatter',x=val,y='FGPer')
    plt.plot(new_x,predictions,c='orange',linewidth=3)
    plt.show()
    
    results.append([val,model.rsquared,model.params[0],model.params[1],model.pvalues[1]])
```

```python
model_formula_params=pd.DataFrame(results)
model_formula_params.columns=model_formula_params.iloc[0]
formula_params=model_formula_params.reindex(model_formula_params.index.drop(0))
formula_params
```

# Model Creation using train_test method in SK Learn

```python
X=data[['TwoPPer','ASTPer','DRBPer','FTPer','G','PER','TOVPer','TRBPer','TSPer','USGPer','VORP','eFGPer']]
target=data['FGPer']
len(target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,target,test_size=0.2,random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression=LinearRegression()
```

```python
linear_regression.fit(X_train,y_train)
```

```python
y_hat_train=linear_regression.predict(X_train)
y_hat_test=linear_regression.predict(X_test)
```

```python
train_residuals = y_hat_train-y_train
test_residuals = y_hat_test-y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_mse=mean_squared_error(y_train,y_hat_train)
test_mse=mean_squared_error(y_test,y_hat_test)
print(train_mse)
print(test_mse)
```
